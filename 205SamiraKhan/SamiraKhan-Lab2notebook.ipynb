```python
#Numerically integrating an ODE.  We will focus on a simple numerical method (Euler's)
#This method is applicable for any first order ODE that can be written as: dy/dt = f(y,t)

#Let's start with a relatively simple ODE:  dN/dt = -a*N  (i.e. exponential decay, with solution
# N(t) = Const.*exp(-a*t)).  This is the radioactive decay probem that we've seen before.

#first we create a function that represents the right hand side of the differential equation.
#Here we pass in the parameter a and the variable N
def ExpDecay(Nval,aval):
    rchange = -aval*Nval
    return rchange

#next we set the parameter value and the initial condition for N (N(0))
a = 5
N = [2000]  #generating a list for values of N, with the initial value, N(0) = 2000, as the first element.
time = [0]  #list of time values, starting with the initial value t=0
dt = 0.02   #The time step we will be taking
Tf = 2     #The final time

#now let's use Euler's methods to advance the x value forward in time
while time[-1] < Tf:   #The while statement.  It loops over the code below until the most recent time in
                       #the time list (i.e. time[-1]) is greater than Tf, the final time.
    N_now = N[-1]       #the current value of N
    t_now = time[-1]    #the current time
    N_next = N_now + ExpDecay(N_now,a)*dt    #Advancing N with Euler's method
    #  Euler's method approximates dN/dt by (N(i+1) - N(i))/(t(i+1) - t(i))
    # Then putting it in the ODE dN/dt = f(N,t)  gives N(i+1) = N(i) + f(N,t)*(t(i+1) - t(i))  (which we implement above)
    t_next = t_now + dt        #advancing the time (important for breaking out of the while loop)
    N.append(N_next)           #adding the updated value of N to the list N
    time.append(t_next)        #adding the update time to the time list
        
#now let's quickly plot N vs. time
plt.plot(time, N)      #this creates the plot
plt.ylabel('N - number of radioactive particles')  #this adds the y-label
plt.xlabel('Time')                                 #this adds the x-label
plt.title('Radioactive Decay')    #also the title of the graph
plt.show()

#After running this cell, try a few different values for a, N[0], dt, and Tf.  
#Comment in a new cell on how the solution (what you see in the graph) changes based on changing these parameters
```

```python
#Say you wanted to know how long it takes for the number of radioactive particles to be reduced by 50% (the half-life)
#Since we don't have an analytical solution, we will have to iterate through the numerical solution to find this time
N_Threshold = N[0]*0.5   #the threshold is 50% of the original
Threshold_Crossed = False    #a boolean value that we will change to True after N crosses N_Threshold
counter = 0                 #a counter that we will increment in the while loop
while not Threshold_Crossed:
    if N[counter] < N_Threshold:
        Threshold_Crossed = True    #change the boolean to True so that we will exit the while loop
        print("The half-life of this sample is ", time[counter], " seconds.")  #print out the time
    else:
        counter += 1     #we need to increment the counter
```
