The Mandelbrot Set

Consider the function $f\left(z\right) = z^2 + c$.  Here, $z$ is a complex number, which we consider to be a point in the complex plane.  $c$ is likewise a complex number.  Starting with $z = 0$, imagine evaluating the function, taking the resulting number, and plugging it back into the function.  Repeating this procedure creates a sequence of complex numbers $\left(0,f\left(0\right),f\left(f\left(0\right)\right),\cdots\right)$, which we can view as a point hopping around on the complex plane.  This sequence has two types of behavior: Bounded, where each point in the sequence is always within a given radius of the origin; and Unbounded, where the point gets farther and farther from the origin with each iteration.  For a given value of $c$, we consider it to be in the Mandelbrot set if the sequence (starting with $z = 0$) is bounded.  So, overall the Mandelbrot set consists of all the points $c$ in the complex plane such that the orbit of $z = 0$ under repeated iteration of the function $f\left(z\right) = z^2 + c$ remains bounded.  It turns out that this set is a fractal, meaning that patterns repeat themselves as we zoom in on the boundary of the set.  

```python
#Now let's create an image of the Mandelbrot Set 
#FYI: Running this code might take a little while (~1 minute), so don't be surprised when you have to wait a bit!
#The code will run faster by reducing the next three variables, but it will result in a less-sharp image
#Increasing these three variables can potentially make the code take a very long time to run (but produce a really nice image)

xpixels = 600  #The number of pixels to use in the real direction (x-direction)
ypixels = 500  #The number of pixels to use in the imaginary direction (y-direction)
#so the total number of pixels in our image is xpixels*ypixels 
#you can reduce this resolution to have the code run faster
iterMax = 500  #the maximium number of iterations to try 
#reducing this also makes the code execute faster, but also makes the image less sharp looking

#the range in which we plot the fractal is a rectangle with the following complex numbers describing the 
#upper left corner point and the lower right corner point.
#The usual domain: upper left point is (-2.1,1.25) and lower right point is (0.9,-1.25)
upperleftpoint = -2.1+1.25j   #Change these to pick a different rectange (i.e to zoom into interesting area)
lowerrightpoint = 0.9-1.25j     #also these
#If you'd like to zoom, try these points: upperleftpoint = -1.1 + 0.6j  and  lowerrightpoint = -0.6 + 0j

diff = lowerrightpoint - upperleftpoint
xrange = abs(diff.real)   #this gets the range in the x direction
yrange = abs(diff.imag)   #this gets the range in the y direction
dx = xrange/(xpixels-1)   #this specifies the distance between two neighboring points in the x direction
dy = yrange/(ypixels-1)   #this specifies the distance between two neighboring points in the y direction

#The next three lines of code are for plotting with color (you don't need to know exactly how they work)
cm = plt.get_cmap('YlOrRd')   #You can change the color scheme ...
#go to https://matplotlib.org/examples/color/colormaps_reference.html  and pick the label to the left of the 
#color scheme that you like best
cNorm  = colors.Normalize(vmin=0, vmax=math.log(iterMax))  #sets the range for a color map
scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=cm)   #this creates the color mapping


MBSim = Image.new("RGB", (xpixels,ypixels))  #This creates a new image (MBSim) to which we will add pixels
#Now we iterate over all of the pixel positions (x and y), and determine if the point z = x+yi is in the set
for y_ind in range(0,ypixels):
    for x_ind in range(0,xpixels):
        c = upperleftpoint + complex(dx*float(x_ind),-dy*float(y_ind))  #Important!  this iterates through a grid
        #of initial positions for the complex number c
        z = 0 + 0j   #the initial value of z
        counter = 0  #this variable tracks how many times we go through the while loop
        
        while abs(z) < 2 and counter < iterMax:  #this keeps going until the point diverges (goes outside of radius 2)
            #or we go past the maximum number of iterations
            z = z**(2) + c  #This is the function that we are iterating (really, it is only this one line)
            #if you want to try out other functions to see what fractal results, modify the above line (RHS)
            counter += 1    #we need to increment the counter to keep from getting stuck in an infinite loop!
        if counter == iterMax:  #the case that the point is bounded
            color_value = (255,255,255)  #this defines a color (white here) with an RGB (Red, Green, Blue) value
        else:  #the case that the point is unbounded
            colorVal = scalarMap.to_rgba(math.log(counter))  #the color is based off of how quickly the point diverged
            #(diverged quickly = smaller value of counter)  This uses the color map that we defined previously
            color_value = (int(colorVal[0]*255),int(colorVal[1]*255),int(colorVal[2]*255))  #this just converts the
            #color representation to a form we can use (colorVal element range from 0-1 continuously,
            # while color_value goes from 0 to 255 descretly)
        MBSim.putpixel( (x_ind,y_ind), color_value)  #this places a pixel of color color_value at 
        #the position (x_ind,y_ind)

fig, ax = plt.subplots(figsize=(14,14))  #this sets the size of our image
ax.imshow(MBSim,extent=[upperleftpoint.real,lowerrightpoint.real,lowerrightpoint.imag,upperleftpoint.imag])
#this shows the image below (and makes the correct axes)
```
